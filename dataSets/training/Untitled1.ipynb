```python
import pandas as pd
import matplotlib as plt
```

```python
train = pd.read_csv("volume(table 6)_training.csv")
test = pd.read_csv("../testing_phase1/volume(table 6)_test1.csv")
```

```python
all_data = pd.concat((train.copy(), test.copy()))
all_data["time"] = all_data["time"].apply(lambda x: pd.Timestamp(x))
all_data["count"] = 1
all_data.index = all_data.time
all_data0 = all_data[(all_data["tollgate_id"] == 3) & (all_data["direction"] == 1)][["time","count"]]
del all_data0["time"]
```

```python
all_data0 = all_data0.resample("20T").sum()
all_data0["time"] = all_data0.index
all_data0["hour"] = all_data0["time"].apply(lambda x: x.hour)
all_data0["minute"] = all_data0["time"].apply(lambda x: x.minute)
del all_data0["time"]
all_data0
```

```python
data_flow = all_data0[(all_data0["hour"] == 7) & (all_data0["minute"] == 0)]["count"]
all_data0["count"].plot()
```
